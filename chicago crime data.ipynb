```python
# import the data from csv

HOUSING_PATH = "datasets/chicago_crime"

import os
import pandas as pd

def load_housing_data(housing_path=HOUSING_PATH):
    csv_path = os.path.join(housing_path, "Chicago_Crimes_2012_to_2017.csv")
    return pd.read_csv(csv_path)

load_housing_data()
```

```python
# time to create a test set, set aside 20% of the data to test on

from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(crimes, crimes["Year"]):
    strat_train_set = crimes.loc[train_index]
    strat_test_set = crimes.loc[test_index]
```

```python
# let's look at proportions of years based on test and train set
crimes["Year"].value_counts() / len(crimes)
```

```python
# the values are so high that the proportions are barely off,
#   but they are there if you look closely enough ;)
strat_train_set["Year"].value_counts() / len(strat_train_set)
```

```python
# time to copy the training data into a new variable that we can
#   mess with. CRIMES NOW EQUALS TRAINING_SET< NOT THE WHOLE THING
crimes = strat_train_set.copy()
```
